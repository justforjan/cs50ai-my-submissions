```python
import pandas as pd
```

```python
df = pd.read_csv('shopping.csv')
```

```python
df.head(10)
```

# Data Transformation
## Text to numeric

```python
df['Month'].unique()
df['VisitorType'].unique()
df['Weekend'].unique()
```

```python
df['Month'] = df['Month'].replace({'Jan': 0, 'Feb': 1, 'Mar': 2, 'Apr': 3, 'May': 4, 'Jun': 5, 'June': 5, 'Jul': 6, 'Aug': 7, 'Sep': 8, 'Oct': 9, 'Nov': 10, 'Dec': 11})

df['VisitorType'] = df['VisitorType'].replace({'Returning_Visitor': 1, 'New_Visitor': 0, 'Other': 0})

df['Weekend'] = df['Weekend'].replace({True: 1, False: 0})

df['Revenue'] = df['Revenue'].replace({True: 1, False: 0})

df.head(10)
```

## Datatypes

```python
df = df.astype({'Administrative': int, 
                'Administrative_Duration': float,
                'Informational': int, 
                'Informational_Duration': float,	
                'ProductRelated': int,	
                'ProductRelated_Duration': float,
                'BounceRates'	: float,
                'ExitRates'	: float,
                'PageValues':	float,
                'SpecialDay':	float,
                'Month'	: int,
                'OperatingSystems':	int,
                'Browser'	: int,
                'Region'	: int,
                'TrafficType':	int,
                'VisitorType':	int,
                'Weekend'	:int,
                'Revenue':int
                })
```
